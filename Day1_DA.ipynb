```python
#Take a number from the user salary if it's More than 50,000 ->10% tax print his final salary
a=int(input("Enter your salary:"))
if(a>=50000):
    t=(a*0.1)
    print("10% tax :{0:.2f} reducted".format(t))
else:
  print("no TAX reduction")
print("Your final salary is ",(a-t))
```

```python
#Get the salary from the user if it's >= 50,000 ->10% tax ,if it's <50000 apply 5% tax if it's more than 70000 apply 12% tax
a=int(input("Enter your salary:"))
if(a<=50000):
    t=a*0.05
    print("5% tax :{0:.2f} reducted".format(t))
elif(a<=70000):
    t=a*0.1
    print("10% tax :{0:.2f} reducted".format(t))
elif(a<=100000):
    t=a*0.12
    print("10% tax :{0:.2f} reducted".format(t))
else:
  print("Promoted")
print("Your final salary is ",(a-t))
```
