#Array Manipulation

```python
#numpy continuation
import numpy as np
```

```python
#Array Manipulation
a = np.linspace(1,11,10,dtype=int)
print(a.shape)
b=a.reshape(5,2)#changes should be saved in a new array
print(b)
```

```python
a = np.linspace(1,11,10,dtype=int)
a.resize(2,5)#Changes the original array
print(a)
```

```python
#Array Operations
a=np.array([12,45,39,65,48,5,23,78,2]).reshape(3,3)
b=np.array([1,65,8,6,4,7,5,7,83]).reshape(3,3)
#We cannot perform arithematic operation on list elements but we can perform on lists
print(f"Subtraction of a and b is {a-b}")
print(f"Addition of a and b is {a+b}")
print(f"One on One Multiplication of a and b is {a+b}")
print(f"Matrix Multiplication of a and b is {a@b}")
```

```python
#min and max function in array
a=np.arange(1,10).reshape(3,3)
#b=np.arange(1,10).reshape(3,3)
print(f"Max of the array is {a.min()}")
print(f"Max of each column is {a.min(axis=0)}")
print(f"Max of each row is {a.min(axis=1)}")
```

```python
a=np.arange(1,10).reshape(3,3)
#b=np.arange(1,10).reshape(3,3)
print(f"Max of the array is {a.max()}")
print(f"Max of each column is {a.max(axis=0)}")
print(f"Max of each row is {a.max(axis=1)}")
```

```python
#Joining arrays(Stacking)
a=np.arange(1,10).reshape(3,3)
b=np.arange(11,20).reshape(3,3)
print(f"Vertical stack on top of another is\n {np.vstack((a,b))} ")
print(f"Horizontal stack one beside of another is\n {np.hstack((a,b))}")
```

```python
a=np.arange(1,10).reshape(3,3)
b=np.arange(11,20).reshape(3,3)
print(f" {np.stack((a,b),axis=0)}")
print("\n\n")
print(f" {np.stack((a,b),axis=1)}")
```

```python
a=np.arange(1,25).reshape(2,3,4)
#b=np.arange(25,51).reshape(2,2,2)
print(a)
print("\nAfter dstack \n")
print(f"{np.dstack((a))}")
#no of groups -> no of column
#no of rows -> no of groups
#no of columns -> no of rows
```

```python
#spliting the arrays
a=np.arange(1,25).reshape(6,4)
print(a,end="\n\n")
print(np.vsplit(a,(1,4,5)))
```

```python
a=np.arange(1,25).reshape(4,6)
print(a,end="\n\n")
print(np.hsplit(a,(2,5)))
```

```python
#arsmax() gives the index of max ele in array
a=np.arange(1,25).reshape(4,6)
print(a.argmax())
```

```python
#arsmax() gives the index of max ele in array
a=np.arange(1,25).reshape(4,6)
print(a)
print(a.argmax(axis=0))#vertical
```

```python
#arsmax() gives the index of max ele in array
a=np.arange(1,25).reshape(4,6)
print(a.argmax(axis=1))#horizontal
```

```python
#arsmax() gives the index of max ele in array
a=np.array([[12,325,65,6,89],[54,36,54,78,87],[32,3,54,67,8]])
print(a)
print(a.argmin(axis=1))#horizontal
```

```python
#arsmax() gives the index of max ele in array
a=np.array([[12,325,65,6,89],[54,36,54,78,87],[32,3,54,67,8]])
print(a)
print(a.argmin(axis=0))#vertical
```

#Statistics

```python
#
a=np.arange(1,10).reshape(3,3)
print(f"Mean of the array is {np.mean(a)}")
print(f"Mean of the array is {np.median(a)}")
print(f"Mean of the array is {np.var(a)}")
print(f"Mean of the array is {np.std(a)}")
```

```python
p = np.pi
print(p)
```

```python
p = np.pi
dp=np.rad2deg(p)
print(dp)
```

```python
a=np.array([p/4,p/2,p/6])
np.deg2rad(a)
```

#Searching

#Algebra

```python
a=np.arange(1,6)
b=np.arange(11,16)
print(a,b,sep='\n')
print(f"Additon of using function is {np.add(a,b)}")
print(f"Additon of using function is {np.subtract(a,b)}")
print(f"Additon of using function is {np.sort(a)}")
print(f"Additon of using function is {np.diff(a)}")
```

```python
#set operations on np arrays
print(f"Additon of using function is {np.setdiff1d(a,b)}")
print(f"Additon of using function is {np.intersect1d(a,b)}")
print(f"Additon of using function is {np.union1d(a,b)}")
```

#Rounding

```python
a=np.array([1.232,12.123,23.5234])
print
print(f"Trucate {np.trunc(a)}")# to Lowest
print(f"Around {np.around(a,2)}")#round off with given decimal points
```

```python
print(f"ceil {np.ceil(a)}")#upper values
print(f"fix {np.fix(a)}")#to lowest similar to trunc
```
