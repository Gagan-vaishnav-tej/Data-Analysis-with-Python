```python
import numpy as np
#ind vs aus live ind has given 298 for aus the runs for every 5 overs is given as follows
#[25,51,84,131,160,189,220,250,267,298]
#aus have completed only 25 overs [15,]
#plot overs vs runs with ind as b aus as #marker for ind is flag,aus is plus
#take overs frome arange function create runs as array
indscore=np.array([25,51,84,131,160,189,220,250,267,298])
ausscore=np.array([15,41,94,110,151])
overs =np.arange(5,51,5)
plt.plot(overs,indscore,label="India",marker=">",color='b',linewidth=3)
plt.plot(overs[:5],ausscore,label="Australia",color='y',marker="+",linewidth=3)
plt.legend()
plt.xlabel("Overs")
plt.ylabel("Runs")
plt.title("Ind vs Aus")
plt.grid()
```
