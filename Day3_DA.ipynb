#Data Cleaning with pandas

```python
df =pd.read_csv("/content/grades_withnulls.csv")
print(df)
```

```python
print(df.isnull())
```

```python
print(df.isnull().sum())
```

```python
#dropna() does not reflect main df by default
#we can overcome this by using inrepalce=TRUE in dropna() function
df.dropna(df)
```

```python
#we can overcome this by using inrepalce=TRUE in dropna() function
mdf =pd.read_csv("/content/grades_withnulls.csv")
mdf.dropna(inplace=True)
print(mdf)
```

```python
df1 =pd.read_csv("/content/grades_withnulls.csv")
df1.fillna(555)
```

```python
print(df1)
```

```python
#fillna() does not reflect main df by default
#we can overcome this by using inrepalce=TRUE in fillna() function
df1.fillna(555,inplace=True)
df1
```

```python
df2 =pd.read_csv("/content/grades_withnulls.csv")
mv = df2['SEM1'].mean()
print(mv)
```

```python
mydf =pd.read_csv("/content/grades_withnulls.csv")
mv = mydf['SEM1'].mean()
mydf.fillna(mv,inplace=True)
mydf
```
